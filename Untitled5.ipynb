```python
import matplotlib.pyplot as plt
import numpy as np
#puntos en la gráfica
x=[0,1,2,3,4,1,2,3,2,1,2,3]
y=[0,0,0,0,0,1,1,1,2,3,3,3]

#graficamos los puntos originales
plt.scatter(x,y,label='Disquette')

plt.xlabel('x')
plt.ylabel('y')
plt.title('rotar el disquette por 45')
plt.legend()

#rotar por num de grados
theta = np.radians(90)
matriz_rotacion = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])
imagen_rotacion = np.dot(matriz_rotacion, np.array([x,y]))

print(f"la matriz de rotacion por 45 grados es {imagen_rotacion}")
#graficar la rotación
plt.scatter(imagen_rotacion[0], imagen_rotacion[1], label='rotacion', marker='x')
plt.show()
```
